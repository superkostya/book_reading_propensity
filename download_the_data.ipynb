# Book Reading Propensity: Downloading the Dataset

```python
import pandas as pd
import gdown
import os
```

```python
DATA_DIR = './data/'
```

```python
file_ids = pd.read_csv(os.path.join(DATA_DIR, 'gdrive_id.csv'))
display(file_ids)
```

```python
file_id_map = dict(zip(file_ids['name'].values, file_ids['id'].values))

def download_by_name(fname, output=None, quiet=False):
    if fname in file_id_map:
        url = 'https://drive.google.com/uc?id='+file_id_map[fname]
        gdown.download(url, output=output, quiet=quiet)
    else:
        print('The file', fname, 'can not be found!')
```

```python
download_by_name('goodreads_interactions_fantasy_paranormal.json.gz')
```
